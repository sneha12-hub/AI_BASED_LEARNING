```python
import google.generativeai as genai
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Configure Gemini AI
genai.configure(api_key="YOUR_GEMINI_API_KEY")
model = genai.GenerativeModel('gemini-pro')

def load_dataset(source):
    """Loads a dataset from seaborn, a CSV file, or a URL."""
    try:
        if source in sns.get_dataset_names():
            return sns.load_dataset(source)
        elif source.endswith('.csv'):
            return pd.read_csv(source)
        else:
            print("Dataset not found! Try a valid CSV file or seaborn dataset name.")
            return None
    except Exception as e:
        print("Error loading dataset:", e)
        return None

def plot_data(request):
    """Generates a seaborn plot based on user input."""
    words = request.lower().split()
    
    dataset_name = None
    plot_type = "scatter"
    x_col, y_col = None, None

    # Detect dataset name
    available_datasets = sns.get_dataset_names()
    for word in words:
        if word in available_datasets or word.endswith('.csv'):
            dataset_name = word
            break

    if not dataset_name:
        print("Please specify a dataset (e.g., 'iris', 'tips', or a CSV file path).")
        return

    # Load dataset
    data = load_dataset(dataset_name)
    if data is None:
        return

    # Detect x and y columns automatically
    numeric_cols = data.select_dtypes(include=np.number).columns
    if len(numeric_cols) >= 2:
        x_col, y_col = numeric_cols[0], numeric_cols[1]
    elif len(numeric_cols) == 1:
        x_col = numeric_cols[0]

    # Detect plot type
    for plot in ["scatter", "histogram", "line", "bar", "box", "heatmap"]:
        if plot in words:
            plot_type = plot
            break

    plt.figure(figsize=(8, 6))

    if plot_type == "scatter" and y_col:
        sns.scatterplot(data=data, x=x_col, y=y_col)
    elif plot_type == "histogram":
        sns.histplot(data[x_col], kde=True, bins=20)
    elif plot_type == "line" and y_col:
        sns.lineplot(data=data, x=x_col, y=y_col)
    elif plot_type == "bar" and y_col:
        sns.barplot(data=data, x=x_col, y=y_col)
    elif plot_type == "box" and y_col:
        sns.boxplot(data=data, x=x_col, y=y_col)
    elif plot_type == "heatmap":
        sns.heatmap(data.corr(), annot=True, cmap="coolwarm")
    else:
        print("Unsupported plot type or not enough numeric columns.")
        return

    plt.title(f"{plot_type.capitalize()} Plot of {dataset_name}")
    plt.show()

def draw_shapes(request):
    """Draws multiple geometric shapes based on user input."""
    words = request.lower().split()
    
    # Default values
    num_circles = 1
    num_rectangles = 1
    num_triangles = 1
    
    # Extract number of shapes (if specified)
    for i, word in enumerate(words):
        if word.isdigit():
            num = int(word)
            if "circle" in words[i + 1:]:
                num_circles = num
            elif "rectangle" in words[i + 1:]:
                num_rectangles = num
            elif "triangle" in words[i + 1:]:
                num_triangles = num
    
    plt.figure(figsize=(6, 6))
    ax = plt.gca()
    
    # Draw Circles
    if "circle" in words:
        for _ in range(num_circles):
            x, y = random.uniform(0.2, 0.8), random.uniform(0.2, 0.8)  # Random positions
            radius = random.uniform(0.1, 0.2)
            circle = plt.Circle((x, y), radius, color=random.choice(['blue', 'red', 'green']), fill=False)
            ax.add_patch(circle)
    
    # Draw Rectangles
    if "rectangle" in words:
        for _ in range(num_rectangles):
            x, y = random.uniform(0.1, 0.7), random.uniform(0.1, 0.7)
            width, height = random.uniform(0.1, 0.3), random.uniform(0.1, 0.3)
            rect = plt.Rectangle((x, y), width, height, color=random.choice(['purple', 'orange']), fill=False)
            ax.add_patch(rect)
    
    # Draw Triangles
    if "triangle" in words:
        for _ in range(num_triangles):
            x1, y1 = random.uniform(0.2, 0.8), random.uniform(0.2, 0.8)
            x2, y2 = x1 + random.uniform(-0.2, 0.2), y1 + random.uniform(-0.2, 0.2)
            x3, y3 = x1 + random.uniform(-0.2, 0.2), y1 + random.uniform(-0.2, 0.2)
            triangle = plt.Polygon([[x1, y1], [x2, y2], [x3, y3]], color=random.choice(['brown', 'cyan']), fill=False)
            ax.add_patch(triangle)
    
    # Set limits and show plot
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    plt.axis('equal')
    plt.title("Geometric Shapes")
    plt.show()


print("Welcome to the AI Chatbot! Type 'exit' to quit.")
while True:
    user_input = input("You: ")

    if user_input.lower() == 'exit':
        print("Goodbye!")
        break

    if any(word in user_input.lower() for word in ["plot", "draw", "graph", "chart"]):
        if any(shape in user_input.lower() for shape in ["circle", "rectangle", "triangle"]):
            draw_shape(user_input)
        else:
            plot_data(user_input)
    else:
        response = model.generate_content(user_input)
        print("Gemini:", response.text)
```
